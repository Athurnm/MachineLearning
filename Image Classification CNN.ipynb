# Inputting Dataset

```python
#Train and Test Directory
train_src_dir="train"
valid_src_dir="validating"
test_src_dir="test"
target_directory="train-class"
valid_target_directory="valid-class"
```

```python
#Making new directory for new classes (train)
import os
from shutil import copyfile

if os.path.isdir(target_directory)==False:
    os.mkdir(target_directory)
if os.path.isdir(target_directory+"/1")==False:    
    os.mkdir(target_directory+"/1")
if os.path.isdir(target_directory+"/2")==False:
    os.mkdir(target_directory+"/2")
if os.path.isdir(target_directory+"/3")==False:
    os.mkdir(target_directory+"/3")
if os.path.isdir(target_directory+"/4")==False:
    os.mkdir(target_directory+"/4")  
```

```python
#Making new directory for new classes (valid)

if os.path.isdir(valid_target_directory)==False:
    os.mkdir(valid_target_directory)
if os.path.isdir(valid_target_directory+"/1")==False:    
    os.mkdir(valid_target_directory+"/1")
if os.path.isdir(valid_target_directory+"/2")==False:
    os.mkdir(valid_target_directory+"/2")
if os.path.isdir(valid_target_directory+"/3")==False:
    os.mkdir(valid_target_directory+"/3")
if os.path.isdir(valid_target_directory+"/4")==False:
    os.mkdir(valid_target_directory+"/4")  
```

```python
#Reading file and enter new file to classes (for training)
train_image_path = []

for filename in os.listdir(train_src_dir):
    if filename.endswith(".jpg"): 
        tempfilename = train_src_dir+"/"+filename
        if os.path.getsize(tempfilename) > 10 :
            #print(os.path.join(directory, filename))
    #        print(filename)       
            filename_str,ext_Str=filename.split(".")
    #        print(filename_str[11])
            oldfilename=train_src_dir+"/"+filename
            newfilename=target_directory+"/"+filename_str[11]+"/"+filename
            print("copy:"+oldfilename+" -> "+newfilename)
            copyfile(oldfilename,newfilename)
    #         if os.path.getsize(newfilename) > 10 :
    #             print(os.path.getsize(newfilename))
            train_image_path.append(os.path.abspath(newfilename))
    #         else :
    #             print("File Invalid")
```

```python
#Reading file and enter new file to classes (for training)
valid_image_path = []

for filename in os.listdir(valid_src_dir):
    if filename.endswith(".jpg"): 
        tempfilename = valid_src_dir+"/"+filename
        if os.path.getsize(tempfilename) > 10 :
            #print(os.path.join(directory, filename))
    #        print(filename)       
            filename_str,ext_Str=filename.split(".")
    #        print(filename_str[11])
            oldfilename=valid_src_dir+"/"+filename
            newfilename=valid_target_directory+"/"+filename_str[11]+"/"+filename
            print("copy:"+oldfilename+" -> "+newfilename)
            copyfile(oldfilename,newfilename)
    #         if os.path.getsize(newfilename) > 10 :
    #             print(os.path.getsize(newfilename))
            valid_image_path.append(os.path.abspath(newfilename))
    #         else :
    #             print("File Invalid")
```

```python
# Importing the Keras libraries and packages
import tensorflow.keras as Keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D
from tensorflow.keras.layers import MaxPooling2D
from tensorflow.keras.layers import Flatten, BatchNormalization
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.preprocessing import image as Image
```

# Deep Learning Classifier

```python
# Creating Model
# Instatiate model
classifier = Sequential()

# Making Convolutional Layer
classifier.add(Conv2D(32, (3, 3), input_shape = (64, 64, 3), padding ='same', activation = 'relu'))
classifier.add(BatchNormalization())

classifier.add(Conv2D(32, kernel_size=(3, 3), activation='relu'))
classifier.add(BatchNormalization())
classifier.add(MaxPooling2D(pool_size=(2, 2)))
classifier.add(Dropout(0.25))

classifier.add(Conv2D(64, kernel_size=(3, 3), activation='relu'))
classifier.add(BatchNormalization())
classifier.add(Dropout(0.25))

classifier.add(Conv2D(128, kernel_size=(3, 3), activation='relu'))
classifier.add(BatchNormalization())
classifier.add(MaxPooling2D(pool_size=(2, 2)))
classifier.add(Dropout(0.25))

classifier.add(Flatten())

# Making output layer
classifier.add(Dense(units = 128, activation = 'relu'))
classifier.add(BatchNormalization())
classifier.add(Dropout(0.5))

classifier.add(Dense(units = 4, activation = 'softmax'))
```

```python
classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
classifier.summary()
```

```python
train_datagen = Image.ImageDataGenerator(rescale = 1./255,
                    shear_range = 0.2,
                    zoom_range = 0.2,
                    vertical_flip = True)
test_datagen = Image.ImageDataGenerator(rescale = 1./255)
training_set = train_datagen.flow_from_directory('train-class',
                target_size = (64, 64),
                batch_size = 32,
                class_mode = 'categorical')

validating_set = train_datagen.flow_from_directory('valid-class',
                target_size = (64, 64),
                batch_size = 32,
                class_mode = 'categorical')

test_set = test_datagen.flow_from_directory('test',
                target_size = (64, 64),
                batch_size = 32,
                class_mode = 'categorical')
```

```python
count_image = len(train_image_path)
print(count_image)
```

```python
from tensorflow.keras.callbacks import EarlyStopping
classifier.fit_generator(training_set,
            steps_per_epoch = count_image,
            epochs = 5, 
            validation_data=validating_set, 
            validation_steps=2000,
            callbacks=[EarlyStopping(monitor='val_loss', patience=3, min_delta=0.0001)])
              
```

```python
# Evaluate accuracy with test data chunk
accr = classifier.evaluate(validating_set)
print('Test set\n  Loss: {:0.3f}\n  Accuracy: {:0.3f}'.format(accr[0],accr[1]))
```

# Shalllow Learning Model

```python
shallow_classifier = Sequential()
shallow_classifier.add(Conv2D(32, (3, 3), input_shape = (64, 64, 3), activation = 'relu'))
shallow_classifier.add(MaxPooling2D(pool_size = (2, 2)))
shallow_classifier.add(Flatten())
shallow_classifier.add(Dense(units = 4, activation = 'softmax'))
```

```python
shallow_classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
shallow_classifier.fit_generator(training_set,
            steps_per_epoch = count_image,
            epochs = 5, 
            validation_data=validating_set, 
            validation_steps=2000,
            callbacks=[EarlyStopping(monitor='val_loss', patience=3, min_delta=0.0001)])
# validation_data = test_set,
# validation_steps = 2000
              
```
